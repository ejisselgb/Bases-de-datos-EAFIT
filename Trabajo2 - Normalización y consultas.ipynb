## Definición del proyecto

Teniendo en cuenta los requirimientos de trabajo anterior y los ajustes de la calificación (la cual se estará subiendo en el transcurso de la semana) se deberá normalizar la base de datos y además se deberán crear los servicios con las consultas que se necesiten para lograr conectar el backend con el frontend y hacer un sitio web de películas funcional.

Las tablas que deberían tener al momento de normalizar son:

- Usuario (información general que tienen todos los usuarios, como cuenta, contraseña)
- Perfil (información personalizada de cada usuario)
- Planes (planes que ofrece la plataforma para suscribirse y ver películas - usar de ejemplo a netflix)
- Suscripciones (aqui se pondrá al usuario y el plan que haya comprado, recuerden poner la fecha de vencimiento y si su suscripcion está activa o no)
- Peliculas
- Series
- Capitulos_Series
- Categorias
- Peliculas_Categorias
- Series_categorias
- Peliculas_Favoritas
- Actores
- Actores_Peliculas
- Pais
- Zona (departamento)
- Ciudad
- Idiomas
- Idiomas_Peliculas
- Idiomas_Series
- Tiempo_Visualización
- Calificacion_Pelicula

La estructura DEBE SER: nombre de las tablas en MAYÚSCULAS (sin excepción) y guión bajo para poner los espacio entre palabras, los campos deben ser en MINÚSCULAS y también tener guión bajo sin son nombres compuestos.

Recuerde crear el modelo completamente modificado y adaptado a las necesidades de la aplicación a construir (Modelo Relacional).

Luego de tener las tablas modificadas, MANTENER LAS ANTERIORES, NO ELIMINAR NADA DE LO CREADO PREVIAMENTE, SI CONSIDERA NECESARIO UN NUEVO ESQUEMA POR FAVOR NOTIFICARLO VÍA TEMAS O CORREO ELECTRONICO PARA ASIGNARLE UNO NUEVO (Esto aplica si los nombres de las tablas existentes pueden generar conflictos con las nuevas tablas normalizadas)


Agregue datos de películas, agregue información real, por ejemplo una imagen existente (en el transcurso de las semana se les entregará un archivo con las rutas de posters de películas que pueden utilizar). Agregue minimo 10 películas incluyendo series y documentales. 15 usuarios y los actores que considere necesario para completar la información del casting de la película o serie.

Cree las siguientes consultas SQL que luego deberán quedar expuestas como servicio para luego poder usar los endpoints en el front de la aplicación.
