### ¿Qué es Dynamo DB?

**DynamoDB** es una base de datos que funciona como servicio en AWS, es muy popular debido a que es fácil agregar información, adicionalmente no tiene una estructura fija pero permite hacer consultas desde una primary key, una sort key o los mismos atributos de la tabla.

Funciones permitidas en Dynamo:

- Scan
- Query
- PutItem
- GetItem
- UpdateItem
- DeleteItem
